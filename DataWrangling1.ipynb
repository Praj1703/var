## Label Encoding

```python
import pandas as pd
```

```python
from sklearn import preprocessing
```

```python
data_nba = pd.read_csv("./datasets/nba.csv")
```

```python
data_nba["Position"].unique()
```

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
data_nba.head()
```

```python
data_nba.dtypes
```

```python
data_nba["Position"] = label_encoder.fit_transform(data_nba["Position"])
```

```python
data_nba["Position"].unique()
```

```python
data_nba.dtypes
```

GPT for query

```python
data_nba["Salary"] = data_nba["Salary"].fillna(data_nba.groupby("Position")["Salary"].transform('mean'))
```

```python
data_nba.isnull().sum()
```

```python
data_nba.isnull().sum()
```

```python
category = pd.cut(data_nba.Age, bins = [19,25,30,35,45], labels = ['A','B','C','D'])
```

```python
data_nba.insert(5, "Age_Group", category)
```

```python
data_nba.head()
```

```python
data_nba.isnull().sum()
```

```python
data_nba= data_nba.drop(columns= ["College"])
```

```python
data_nba.isnull().sum()
```

```python
data_nba["Age_Group"].value_counts()
```

```python
data_nba["Age_Group"].unique()
```

```python
data_nba["Age_Group"]= data_nba["Age_Group"].fillna('A')
```

```python
data_nba.isnull().sum()
```

```python
data_nba.shape
```

```python
data_nba = data_nba.dropna(axis=1, how='all')
```
