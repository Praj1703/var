```python
# Create Academic Performance Dataset

data = {
    'Name': ['Amit', 'Riya', 'Karan', 'Sneha', 'Rahul',
             'Pooja', 'Vikas', 'Neha', 'Arjun', 'Priya'],
    
    'Maths': [85, 90, np.nan, 45, 120,
              78, 88, 92, 35, 80],
    
    'Science': [80, 85, 70, np.nan, 95,
                76, 89, 91, 40, 84],
    
    'English': [78, 88, 65, 50, 98,
                74, 90, 93, 42, -5],
    
    'Attendance': [90, 95, 85, 60, 110,
                   80, 88, np.nan, 55, 92]
}

df = pd.DataFrame(data)

print("Original Dataset:\n")
df
```

```python
# Detect outliers using IQR method

Q1 = df['Maths'].quantile(0.25)

Q3 = df['Maths'].quantile(0.75)

IQR = Q3 - Q1

print("Q1 =", Q1)
print("Q3 =", Q3)
print("IQR =", IQR)
```

```python
# Define lower and upper limits

lower_limit = Q1 - 1.5 * IQR

upper_limit = Q3 + 1.5 * IQR

print("Lower Limit =", lower_limit)
print("Upper Limit =", upper_limit)
```

```python
# Remove outliers

df_no_outliers = df[(df['Maths'] >= lower_limit) &
                    (df['Maths'] <= upper_limit)]

print("Dataset after removing outliers:\n")

df_no_outliers
```

```python
# Apply log transformation

df_no_outliers['Log_Maths'] = np.log(df_no_outliers['Maths'])

print("Dataset after Log Transformation:\n")

df_no_outliers[['Maths', 'Log_Maths']]
```
