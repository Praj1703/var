```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
sns.histplot(x = 'sepal_length', data = df)
```

```python
sns.boxplot(data = df)
```

```python
q1 = df['sepal_width'].quantile(0.25)
q3 = df['sepal_width'].quantile(0.75)
```

```python
iqr = q3-q1
```

```python
lb = q1 - 1.5 * iqr
```

```python
up = q3 + 1.5 * iqr
```

```python
outliers = df[(df['sepal_width'] < lb) | (df['sepal_width'] > up)]
```
